```python
import plotly.express as px
import plotly.graph_objects as go
```

$y_{t}=5t\left[0.2969{\sqrt {x}}-0.1260x-0.3516x^{2}+0.2843x^{3}-0.1015x^{4}\right]$

```python
#@title Parameters
xx =  20#@param {type:"number"}
c = 1 #@param {type:"number"}
N = 300 #@param {type:"number"}

```

```python
t = xx/100

def y(x):
  return 5*t*(0.2969*(x**0.5) - 0.126*x - 0.3516*(x**2) + 0.2843*(x**3) - 0.1015*(x**4) )

def arange(start,end,N):
  step = (end-start)/(N-1)
  return [start + i*step for i in range(N) ]

xtab = []
y_pos = []
y_neg = []

for x in arange(0,1.0,N):

  xtab.append(x*c)
  y_ = y(x)*c
  y_pos.append(y_)
  y_neg.append(-y_)

xtab = xtab + xtab[::-1]
ytab = y_pos + y_neg[::-1]
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=xtab, y=ytab,
                    mode='lines',
                    name='NACA00' + str(xx)))
fig.update_yaxes(
    scaleanchor = "x",
    scaleratio = 1,
  )
fig.show()
```
